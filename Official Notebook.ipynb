# Dataset Description 

The dataset Commission of Filipino Overseas: Statistical Profile Of EVP Participants [2007-2020] is available on the National Government Portal's Open Data Philippines website. Published by the Commission on Filipinos Overseas (CFO), the dataset was retrieved on February 8, 2023 from the CSO's website. The statistical data used to form the instances were from the Exchange Visitor Program's (EVP) applicants' personal information.
